```python
Keyword_pattern=re.compile("[F|f]arada(y|ic)\'?\s?[E|e]fficienc(y|ies)|\sFE\s|\sFECO\s")
###(Not HClO4)XX.XX mV|XX.XX, | XX.XX and#####
Value_Keyword_pattern=re.compile("\d+\s?%|\d+\.\d+\s?%|\d+\s?\+\-|\d+\.\d+\s?\+\-")
```

```python
###Clean the marks .@/-x###
def clean_mark(target_list):
#     print(target_list)
    try:
        for i in range(0,len(target_list)):
            ###remove chemical value e.g. Ir(IV)###
            target_list[i]=re.sub('\(V?[I]+V?\)','',target_list[i])
            ###replacing special specifiers###
            target_list[i]=target_list[i].replace('/',"")
            target_list[i]=target_list[i].replace('@',"")
            target_list[i]=target_list[i].replace('-',"")
            target_list[i]=target_list[i].replace('x',"")
            target_list[i]=target_list[i].replace('y',"")
            target_list[i]=target_list[i].replace('X',"")
            
    except:
        pass
    return target_list
```

```python
def Compare_whether_same_element_exist(list_a,list_b):
    Judge=False
    for each_a in list_a:
        if each_a in list_b:
            Judge=Judge or True
        else:
            Judge=Judge or False
    return Judge
```

```python
def Check_and_supplement_element(key_word_list,Extracted,Element_list,Element):
    if Compare_whether_same_element_exist(key_word_list,Extracted)==True:
        Element_list.append(Element)
    return Element_list
```

```python
def standardize_element(Element_list,number_elements):
    return_list=copy.deepcopy(Element_list)
    if len(Element_list)<number_elements:
        for i in range(0,number_elements-len(Element_list)):
            return_list.append('None')
    else:
        return_list=return_list[0:number_elements]
    return return_list
```

```python
########Material Speciality##########
def Find_Material(Pattern,text):
    return_list=[]
    Judge=0
    Result=re.findall(Pattern,text,re.IGNORECASE)
    if len(Result)>=1:
        Judge=1
    return_list.append(Judge)
    return return_list
```

```python
###Find the key sentence from the text for further mining###
def Find_Target_Sentence_From_Text(text,Keyword_pattern):
    sentence_list=text.split('. ')
    FE_found_sentence_list=[]
    for each_sentence in sentence_list:
        try:
            FE_index_list=[m.start(0) for m in re.finditer(Keyword_pattern, each_sentence)]
            if len(FE_index_list)>=1:
                FE_found_sentence_list.append(each_sentence)
        except:
            print('error!')
    return FE_found_sentence_list
```

```python
###Turn extracted voltage list like [' 15 mV', ' 512 mV'] into [15.0, 512.0]###
def extract_value(target_list):
#     print(target_list)
    try:
        for i in range(0,len(target_list)):
            target_list[i]=float(target_list[i].strip(' ,%and+-'))
    except:
        pass
    return target_list
```

```python
#############Check_Irregular###############
def Check_Irregular(FE_values_list,FE_index_list,Low_Value_Boundary,High_Value_Boundary,FE_Multiple_Boundary,FE_Keyword_Boundary):
    ###Check if there is irregular high value like OER FE in the same sentence### 
    Irregular_Abnormal_Value=0
    for each_founded_FE_value in FE_values_list:
        if each_founded_FE_value>High_Value_Boundary or each_founded_FE_value<Low_Value_Boundary:
#             print('Irregular value not in range found!, value is',each_founded_FE_value)
            Irregular_Abnormal_Value=1
    ###Check if multi values found### 
    Irregular_Multiple_Value=0
    if len(FE_values_list)>FE_Multiple_Boundary:
       Irregular_Multiple_Value=1
    ###Check if multi keywords found### 
    Irregular_Multiple_Keywords=0
    if len(FE_index_list)>FE_Keyword_Boundary:
       Irregular_Multiple_Keywords=1
    return Irregular_Abnormal_Value,Irregular_Multiple_Value,Irregular_Multiple_Keywords
```

```python
def Extracting_target_values(Keyword_pattern,Value_Keyword_pattern,Target_Sentence,Low_Value_Boundary,High_Value_Boundary,DOI):
    FE_index_list=[m.start(0) for m in re.finditer(Keyword_pattern, Target_Sentence)]
    print('Mention of FE:',len(FE_index_list),'times')
    try:
        founded_values=re.findall(Value_Keyword_pattern,Target_Sentence)
        ###Remove the units###
        FE_values_list=extract_value(founded_values)
        ###Even though some FE times values could be found### 
        final_FE_value_single=max(FE_values_list)
        ###Check Irregular###
        Irregular_Abnormal_Value,Irregular_Multiple_Value,Irregular_Multiple_Keywords=Check_Irregular(FE_values_list,
        FE_index_list,Low_Value_Boundary=Low_Value_Boundary,
        High_Value_Boundary=High_Value_Boundary,FE_Multiple_Boundary=1,
        FE_Keyword_Boundary=1)
        ###Recording###
        Bug_Note='Normal, FE value found.'
        Final_FE=final_FE_value_single
        Remark_Sentence=Target_Sentence
        Bug_Note,Irregular_Abnormal_Value,Irregular_Multiple_Value=Bug_Note,Irregular_Abnormal_Value,Irregular_Multiple_Value
        Irregular_Multiple_Keywords=Irregular_Multiple_Keywords
    except:
        print('Error in finding FE values.',DOI)
        Bug_Note='No FE.value found'+DOI
        ###Recording###
        Final_FE='NaN'
        Remark_Sentence=Target_Sentence
        Bug_Note,Irregular_Abnormal_Value,Irregular_Multiple_Value=Bug_Note,'NaN','NaN'
        Irregular_Multiple_Keywords='NaN'
    
    
    return Final_FE,Remark_Sentence,Bug_Note,Irregular_Abnormal_Value,Irregular_Multiple_Value,Irregular_Multiple_Keywords
```
